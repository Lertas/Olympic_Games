```python
#Malfa Ilia Aikaterini, Lertas George

import numpy as np
import pandas as pd
import statistics

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pylab as pl
```

### Task1

```python
#load data
task1 = pd.read_csv('task1_output',sep='/',names=['ID','Name','Sex','Gold'])
task1
```

```python
#In this plot we can see that we have more than double men with gold medal than women
#Of course because in the past there were fewer female entries
plt.title('Countplot with the Sex of athletes with gold Medals')
sns.countplot(x='Sex',data=task1);
plt.savefig('Task1_plot1.jpeg')
```

```python
#In this plot we can see that we have over that 1.2 gold medals per 1 man
#but there is near to 1.4 gold medals per 1 woman
#which means most different men win a gold medal that women
sns.factorplot(x='Sex',y='Gold',data=task1,kind='bar');
plt.savefig('Task1_plot2.jpeg')
```

```python
#As we can see from the boxplot, most of men have one gold medal
#and from women mostof them have 1 or 2 or 3 gold medals
plt.title('Boxplot with the Sex of athletes with gold Medals')
sns.boxplot(x='Sex', y='Gold',data=task1, palette='rainbow');
plt.savefig('Task1_plot3.jpeg')
```

```python
task1.sort_values(by=['Gold'],ascending=False).head(20)
```

```python
#Here we have a barplot with Athletes with gold medals and hue the sex attribute
#We can see that Michael Fred Phelps, II has a long distance from the second in relation to the gold medals 
d=task1.sort_values(by=['Gold'],ascending=False).head(10)
plt.figure(figsize=(30,5))
plt.title('Countplot with athletes with most gold medals')
sns.barplot(x='Name',y='Gold',data=d,hue='Sex');
plt.savefig('Task1_plot4.jpeg')
```

### Task3

```python
#load data of task3
task3 = pd.read_csv('task3_output',sep='/',names=['Games','Team','NOC','WomenCount','Prominent_Sport'])
task3
```

```python
#As we can see here USA are the team with the most women participations per Olympics
plt.figure(figsize=(20,5))
plt.title('Countplot with the Team of women participations')
sns.countplot(x='Team',data=task3);
plt.savefig('Task3_plot1.jpeg')
```

```python
#Here we have the result at the most common value
task3.Team.mode()
```

```python
#As we can see here Athletics are the Prominent Sport with the most women participations per Olympics
plt.figure(figsize=(20,5))
plt.title('Countplot with the Prominent_Sport of women participations')
sns.countplot(x='Prominent_Sport',data=task3);
plt.savefig('Task3_plot2.jpeg')
```

```python
#Here we have the result at the most common value
task3.Prominent_Sport.mode()
```

```python
#WomenCount is a Ratio-scaled attribute, so we can create a boxplot
plt.figure(figsize=(3,6))
plt.title('Boxplot with the WomenCount of women participations')
sns.boxplot(y='WomenCount',data=task3);
plt.savefig('Task3_plot3.jpeg')
```

```python
#Here we have the results from the WomenCount attribute
task3.describe()
```

```python
#Here we have boxplots participations per Team. 
plt.figure(figsize=(20,5))
plt.title('Boxplot with the WomenCount of women participations per Team')
sns.boxplot(x='Team', y='WomenCount',data=task3, palette='rainbow');
plt.savefig('Task3_plot4.jpeg')
```

```python
#Here we have the results from the WomenCount attribute per Games, same results with as the boxplot above
task3.groupby('Games').describe()
```

```python
#Here we have boxplots with women participations per Olympic Games
#Also, we can see that over the years we have more participations 
#and in Winter Games we have less participations than Summer
plt.figure(figsize=(70,20))
plt.title('Boxplot with the WomenCount of women participations per Games')
sns.boxplot(x='Games', y='WomenCount',data=task3, palette='rainbow');
plt.savefig('Task3_plot5.jpeg')
```

```python
#Here we have the results from the WomenCount attribute per Team, same results with as the boxplot above
task3.groupby('Team').describe()
```
