```python
#load data
df = pd.read_csv('athlete_events.csv',sep=',')
df
```

```python
#Find in the Team attribute values which are double
df[df['Team'].str.find('/')>0].Team
```

```python
#NOC countries with two values in Team attribute
df[df['Team'].str.find('/')>0].NOC.unique()
```

```python
#with find(), we find Teams with 2 values and replace it with the NOC team
for i,j in (df[df['Team'].str.find('/')>0].Team).items():
    if df.iloc[i].NOC == 'DEN':
        df.iloc[i, df.columns.get_loc('Team')] = 'Denmark'
    elif df.iloc[i].NOC =='ITA':
        df.iloc[i, df.columns.get_loc('Team')] = 'Italy'
    elif df.iloc[i].NOC =='GBR':
        df.iloc[i, df.columns.get_loc('Team')] = 'Great Britain'
    elif df.iloc[i].NOC =='USA':
        df.iloc[i, df.columns.get_loc('Team')] = 'United States'
    elif df.iloc[i].NOC =='FRA':
        df.iloc[i, df.columns.get_loc('Team')] = 'France'
    elif df.iloc[i].NOC =='HUN':
        df.iloc[i, df.columns.get_loc('Team')] = 'Hungary'
    elif df.iloc[i].NOC =='AUS':
        df.iloc[i, df.columns.get_loc('Team')] = 'Australia'
    elif df.iloc[i].NOC =='GER':
        df.iloc[i, df.columns.get_loc('Team')] = 'Germany'
    elif df.iloc[i].NOC =='SWE':
        df.iloc[i, df.columns.get_loc('Team')] = 'Sweden'
    elif df.iloc[i].NOC =='BOH':
        df.iloc[i, df.columns.get_loc('Team')] = 'Bohemia'
```

```python
#we find teams with -1, -2, -3, -4 in last two letters and we remove these last characters
for i,j in df['Team'].items():
    if j[-2:]== '-1' or j[-2:]== '-2' or j[-2:]== '-3' or j[-2:]== '-4':
        df.iloc[i, df.columns.get_loc('Team')] = j[:-2]
```

```python
#after searching, we get this csv with NOC and the name of each noc as region attribute
df2 = pd.read_csv('noc_regions.csv',sep=',')
df2
```

```python
#merge the two dataframes with NOC 
result = pd.merge(df, df2, on="NOC")
```

```python
#replace Teams values with region
result = result.assign(Team=result['region'])
```

```python
#drop the columns tha we do not need
result = result.drop(columns=['region', 'notes'])
```

```python
#we check how many nan values we have
result.isnull().sum()
```

```python
#export csv
from tkinter import filedialog

#drop duplicates and replace nan with zeros
result = result.drop_duplicates()
result = result.fillna(0)

export_file_path = filedialog.asksaveasfilename(defaultextension='.tsv')
result.to_csv (export_file_path, index = False, header=True, sep='/')
```
